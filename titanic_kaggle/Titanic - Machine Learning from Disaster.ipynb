### Verificação de valores inconsistentes ou Faltantes

```python
# Não existe idades menores que 0.
titanic_train.loc[titanic_train['Age'] < 0] 
```

```python
titanic_train.isnull().sum()
```

```python
# Localização das idades faltantes
titanic_train.loc[pd.isnull(titanic_train['Age'])]
```

```python
# Tratamento de valores faltantes, subistituindo as idades faltantes pela média
titanic_train['Age'].fillna(titanic_train['Age'].mean(), inplace = True)
titanic_teste['Age'].fillna(titanic_teste['Age'].mean(), inplace = True)
```

```python
titanic_train['Embarked'].mode()[0]
```

```python
titanic_train['Embarked'].fillna(titanic_train['Embarked'].mode()[0], inplace = True)
```
